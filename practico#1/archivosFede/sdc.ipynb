```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

f = np.array([70,80,90,100,110,120,130,140,150,160,170,180,190,200,210,220,230,240,250])
t = np.array([17644,15438,13723,12351,11228,10293,9501,8821,8234,7719,7265,6862,6500,6174,5882,5614,5370,5146,4940])

def func(x, a, b, c):
    return a * x**2 + b * x + c

popt, pcov = curve_fit(func, f, t)

plt.figure(figsize=(10,5))

plt.plot(f, t, 'bo', label='datos originales')
plt.plot(f, func(f, *popt), 'r-', label='curva ajustada: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))

plt.ylabel('Tiempo de ejecución [ms]')
plt.xlabel('Frecuencia del clock [MHz]')
plt.title('Tiempo de ejecución vs frecuencia de clock')
plt.legend()
plt.show()
```
